# Structured Outputs

In LangChain, **structured output** refers to the practice of having language models return
responses in a **well-defined data format** (such as JSON) instead of free-form text.  
This makes the model output **easier to parse, validate, and use programmatically** in
applications. This data can be easily integrated with other system.

## TypedDict in Python

**TypedDict** is a way to define a dictionary in Python where you specify what keys and values  
should exist. It helps ensure that your dictionary follows a specific structure.

### Why Use TypedDict?
- Specifies required keys and their value types
- Improves code readability and IDE support
- Provides static type checking
- Does **not** perform runtime validation (type-hint only)

### Types of TypedDict

- **Simple TypedDict:**  
  Defines keys with their expected value types

- **Annotated TypedDict:**  
  Uses `Annotated` to attach metadata or descriptions to fields

- **Literal:**  
  Restricts values to a fixed set of allowed options

## Pydantic in Python

**Pydantic** is a data validation and data parsing library for Python. It ensures that the data you
work with is correct, structured, and type-safe.
